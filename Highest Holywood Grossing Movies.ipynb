```python
# import library
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# read in the data
df = pd.read_csv(r'C:\Users\User\Desktop\move\Highest Holywood Grossing Movies.csv ')
```

```python
# look at the data

df.head()
```

```python
# missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
# presentation by world profit
df.sort_values(by=['World Sales (in $)'], inplace=False, ascending=False)
```

```python
# scatter diagram with earnings in domestic and international cinemas

plt.scatter(x=df['Domestic Sales (in $)'], y=df['International Sales (in $)'])

plt.title('earnings in domestic and international cinemas')

plt.xlabel('domestic earnings')

plt.ylabel('international earnings')

plt.show()
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# Now we look at those that have a high correlation (> 0.5)
strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```

```python
# Looking at the top 15 companies by gross revenue
CompanyGrossSum = df.groupby('Distributor')[["World Sales (in $)"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('World Sales (in $)', ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['World Sales (in $)'].astype('int64') 

CompanyGrossSumSorted
```
